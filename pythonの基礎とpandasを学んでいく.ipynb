```python
#ダミー変数か
pd.get_dummies(df)
```

```python
#相関係数
df[["survived", "pclass", "age", "sibsp", "parch", "adult_male", "alone"]].corr()
```

```python
 df = df.reset_index()
 df
```

```python
df_a = df[["index","survived"]]
df_b = df[["index","pclass"]]
```

```python
#結合　indexをkeyにしてinner join
pd.merge(df_a, df_b, on="index", how="inner")
```

```python
#１が横の結合　０が縦の結合
pd.concat([df_a, df_b],axis= 1)
```
