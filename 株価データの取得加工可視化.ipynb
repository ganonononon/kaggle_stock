```python
#移動平均線を作る
#min_periodsは最低期間が1日でも表示できるようにするもの
ticker = "AMZN"
df = yf.download(ticker, start="2022-01-01", end="2023-12-31", interval="1d")

price = df["Close"]
span = 5

df["MA_for_5_days"] = price.rolling(window=span, min_periods=1).mean()

df
```

```python
import matplotlib.pyplot as plt
df_ma = df[["Close", "MA_for_5_days"]]

plt.figure(figsize=(15,5))
plt.plot(df_ma.index, df_ma)
plt.legend(df_ma)
plt.show()
```

```python
span = 30
df["MA_for_30_days"] = price.rolling(window=span, min_periods=1).mean()
df_ma = df[["Close", "MA_for_30_days"]]

plt.figure(figsize=(15,5))
plt.plot(df_ma.index, df_ma)
plt.legend(df_ma)
plt.show()
```

```python
#標準偏差
#赤いゾーンを抜けたところが買い目だが、そのあと下がることもあるので注意
span = 30
df["MA_for_30_days"] = price.rolling(window=span, min_periods=1).mean()
df["std"] = price.rolling(window=span, min_periods=1).std()

df["-1d"] = df["MA_for_30_days"] - 1 * df["std"]
df["+1d"] = df["MA_for_30_days"] + 1 * df["std"]
df["-2d"] = df["MA_for_30_days"] - 2 * df["std"]
df["+2d"] = df["MA_for_30_days"] + 2 * df["std"]



plt.figure(figsize=(15,5))
plt.title(ticker + "trend analysis")
plt.plot(df["Close"], color="blue")
plt.plot(df["-1d"], color="red")
plt.plot(df["+1d"], color="red")
plt.plot(df["-2d"], color="green")
plt.plot(df["+2d"], color="green")

plt.fill_between(df.index, df["-1d"], df["+1d"], color="red", alpha=0.5)
plt.fill_between(df.index, df["-2d"], df["+2d"], color="lightgreen", alpha=0.5)
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend(["Close","-1d","+1d", "-2d","+2d"], loc="upper right")
plt.show()
```
